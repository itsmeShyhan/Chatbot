# Part A - Problem Domain Review

Problem Domain Review
Chatbots is when machines are used to mimic human-like conversations. This can be efficient and cost-effective in many ways: Cybersecurity, web analytics, task management, financial advice, customer support, etc. One such example for customer support is the Whatsapp customer service accounts of service providers like Dialog, Mobitel, Hutch, etc in Sri Lanka. Other use cases include product recommendation, screening job candidate. Furthermore, chatbots are also used to summarize news and other articles. (Sanjaya et al., 2023, pp. 432-438)

Chatbots can be developed in two ways depending on the type of business. Large businesses like Ali Baba develop chatbots from scratch which are highly customized solutions. On the other hand, small businesses can develop chatbots using platforms that provide built-in tools and techniques to customize such as Google Dialog Flow, tensor flow, etc. The main AI technique used is Natural Language Processing: helps understand and/or remember the context of the chat as well the sentiment of the human user that interacts with the chatbot. This can lead to the program delivering an appropriate response to the user based on their query. (Sanjaya et al., 2023, pp. 432-438). Some of the main NLP techniques used are tokenization, bag of words, SVM, Deep Learning, etc. Despite all of these methods, understanding complex emotion is still a challenge for chatbots.

AI Chatbots are also used in education in the form of generative AI chatbots like ChatGPT, Gemini etc. It helps both the student and tutors in a wide variety of ways. The three main benefits provided for the students are accessibility and accessibility. Accessibility is on-demand learning where the chatbots provide detailed explanations and feedbacks at anytime and any place with regards to the convenience of the student. Last but not least, Adaptability is when the chatbot provides responses that tailors to different students and how they understand and grasp different concepts. On the other hand, advantages for the tutor include saving time where tasks like grading, feedback, explanations can be provided by AI chatbots for an assessment or lecture. (Labadze, Grigolia, and Machaidze, 2023) Example of such chatbots include Jill Watson &Genie, Hubert, MOOC EdX, etc. (Park, Lee and Song, 2020).

They are also used in healthcare field where the patients can ask questions on their health and report, schedule appointments to meet doctors, reminders on health benefits, etc. All of these can potentially help improve the health of the patient as well as improving the knowledge of the patient. Future trends include the usage of AI Chatbots to provide diagnosis based on medical history and symptoms. (Adamopoulou and Moussiades, 2020).

Chatbots are also used in e-commerce websites like Sephora where they are used to recommend and order products. In addition to this, it also provides “personalized beauty tips, reviews, etc.” Another such example is Spring Bot which “allows you to purchase products directly and provides answers to receipts, shipping information, and other purchasing questions.” (Park, Lee and Song, 2020)

Chatbots are also used in data analysis, operator support, etc. Once such example is the Solvvy AI chatbot can is used to provide insights and answer queries for customers in addition to its 24/7 availability which enhances customer experience as well.
(Park, Lee and Song, 2020)


# Part B - Compare and Evaluate AI Techniques
Part B – Compare and Evaluate AI Techniques

There are different AI techniques that can be used for chatbots. The three main techniques are RNN, CNN, Feedforward Neural Network. RNN Model like Seq2Seq has advantages and disadvantages as well. RNNs work well for chatbots as they are able to remember a moderate length of sequences. This is crucial for ChatBot as remembering details can help in maintaining the context of the chat.(Bhagwath, 2018). But Basic RNNs struggle to work with long sequence of data as they tend to “forget” information from earlier parts of the sequence. This can be a problem for a chatbot where it is possible for a long conversation such as in generative AI chatbots like ChatGPT, Gemini, etc. (Nithuna and Laseena, 2020)

CNNs are primarily used for Image Recognition but it can also be used for text generation and speech recognition as well. CNNs involve using a convolution layer which “reduces the complexity of the matrix”, hence it allowing working with complex data and particularly suited for complex data. On the other hand, Feedforward Neural Network, also known as multi-layer perceptron has one input layer, one layer and at least one hidden layer. This is primarily used for speech recognition and translation part of the NLP. The disadvantage of Feedforward Neural Network is it is too simple for complex data with contrast to the previous technique which works with complex data by having a convolution layer to compress the data. (Bhagwath, 2018)

By considering all three of these techniques, Feedforward Neural Network is the best technique suited for this particular project and scope. This is due to the simplistic nature of this technique which aligns with the goals of this project, which is to allow customers to get basic information on a restaurant like opening hours, payment mode available, menu items available by using the chatbot. Therefore, a simple restaurant information chatbot is well suited to be developed using feed forward neural network.

# NLP Techniques

```python
def tokenizer(text):
  # print(word_tokenize(text))
  return word_tokenize(text)


def stemming(text):
  porter = PorterStemmer()
  stemmedWords = []
  # print(porter.stem(text[0]))
  # for i in text:
  #   # print(porter.stem(text[i]))
  #   stemmedWords.append(porter.stem(i))
  # print(stemmedWords)
  return porter.stem(text)



def bag_of_words(tokenized_sentence, all_words):
  # Stemming the already tokenized words
  stemmed_sentence = [stemming(w) for w in tokenized_sentence]


  # Creating an array of bag words based on the length of all words in intents
  bag = np.zeros(len(all_words), dtype=np.float32)
  for idx, w in enumerate(all_words):
    if w in stemmed_sentence:
      bag[idx] = 1.0
  return bag



```

```python
# Loading the data
data = '''
{
  "intents": [
    {
      "tag": "greeting",
      "patterns": [
        "Hi",
        "Hi, there",
        "Hey there",
        "Yo",
        "What's up?",
        "Hello there",
        "Hey",
        "How are you",
        "Is anyone there?",
        "Hello",
        "wazzup",
        "helloooooo"

      ],
      "responses": [
        "Hey :-)",
        "Hello, thanks for visiting Al Mass Bot. How can I help you?",
        "Hi there, what can I do for you?",
        "Hi there, how can I help?",
        "Hi there! What can I do for you?",
        "Hey, how's it going? What's up?",
        "Ask away.",
        "I'm here to help. What do you need assistance with?",
        "How can I help you today?",
        "Hello! What would you like to know?",
        "Hi there! How can I help you today?",
        "What brings you here?",
        "Hello! What can I do for you?"
      ]
    },
    {
      "tag": "goodbye",
      "patterns": ["Bye", "See you later", "Goodbye"],
      "responses": [
        "See you later, thanks for visiting",
        "Have a nice day",
        "Bye! Come back again soon."
      ]
    },
    {
      "tag": "thanks",
      "patterns": ["Thanks", "Thank you", "That's helpful", "Thank's a lot!"],
      "responses": ["Happy to help!", "Any time!", "My pleasure"]
    },
    {
      "tag": "menu",
      "patterns": [
        "Which items do you have?",
        "What kinds of items are there?",
        "What do you sell?",
        "What items are available?",
        "What products do you offer?",
        "What's on your menu?",
        "What do you have in stock?",
        "What items are currently available for purchase?",
        "What is your product selection?",
        "What are your offerings?",
        "What items do you carry?",
        "What's in your inventory?",
        "What are your available options?",
        "What can I choose from?",
        "What items are you selling?",
        "What do you have for sale?",
        "What are your current items?",
        "What products are you featuring?",
        "What's on the list?",
        "What are my choices?",
        "What items do you provide?",
        "What are your available products?",
        "Do you have a list of items?"
      ],
      "responses": [
        "We sell fried rice, naan rotti, rice and curry, Mixed Fruit Juice",
        "We have fried rice, naan rotti, rice and curry, Mixed Fruit Juice"
      ]
    },
    {
      "tag": "price",
      "patterns": [
        "What are the prices like here?",
        "How much does food typically cost?",
        "What is the average price per meal?",
        "What's the price range for your dishes?",
        "Are your prices affordable?",
        "What is the typical cost of a meal?",
        "How much can I expect to pay for dinner?",
        "What is the price range for your menu items?",
        "Are your prices reasonable?",
        "How much does it usually cost to eat here?",
        "What is the price point of your restaurant?",
        "Is this a fine dining restaurant?",
        "Is this a casual dining restaurant?",
        "Is this a budget-friendly restaurant?",
        "What is the overall price level?"
      ],
      "responses": [
        "Fried Rice: Rs. 800; Fruit Juice: Rs. 400; Rice and Curry: Rs. 900",
        "Fried Rice - Rs. 800, Fruit Juice - Rs. 400, Rice and Curry - Rs. 900",
        "The prices are: Fried Rice - Rs. 800, Fruit Juice - Rs. 400, Rice and Curry - Rs. 900",
        "Here are the prices: Fried Rice - Rs. 800, Fruit Juice - Rs. 400, Rice and Curry - Rs. 900",
        "The menu prices are as follows: Fried Rice - Rs. 800, Fruit Juice - Rs. 400, Rice and Curry - Rs. 900",
        "Fried Rice is priced at Rs. 800, Fruit Juice at Rs. 400, and Rice and Curry at Rs. 900.",
        "You can find Fried Rice for Rs. 800, Fruit Juice for Rs. 400, and Rice and Curry for Rs. 900.",
        "The cost of Fried Rice is Rs. 800, Fruit Juice is Rs. 400, and Rice and Curry is Rs. 900."
      ]
    },
    {
      "tag": "payments",
      "patterns": [
        "Do you take credit cards?",
        "Do you accept Mastercard?",
        "Can I pay with Paypal?",
        "Are you cash only?",
        "Let me know the payments methods available"
      ],
      "responses": [
        "We accept VISA, Mastercard, Paypal and Cash",
        "We accept most major credit cards, Paypal and Cash"
      ]
    },
    {
      "tag": "delivery",
      "patterns": [
        "How long does delivery take?",
        "How long does shipping take?",
        "When do I get my delivery?",
        "What is the estimated delivery time?",
        "When can I expect my order to arrive?",
        "How quickly will my order be delivered?",
        "What is the delivery timeframe?",
        "How long does shipping usually take?",
        "What is the typical delivery window?",
        "When will my order be delivered?",
        "How soon can I expect to receive my order?",
        "What is the delivery turnaround time?",
        "How long does it take for you to deliver?",
        "What's the delivery schedule like?",
        "How long will it take to get my order?",
        "What is the delivery timeframe for this item?",
        "When can I expect to have my order?",
        "What is the delivery window for this order?",
        "How long will it be before I receive my order?",
        "What is the estimated delivery date?",
        "Can you give me an estimated delivery time?",
        "How long does shipping typically take?",
        "What is the normal delivery time?"
      ],
      "responses": [
        "Delivery takes atleast 4 hrs",
        "It will take minimum 4 hrs",
        "4hrs minimum to deliver"
      ]
    },
    {
      "tag": "opening",
      "patterns": [
        "What time are you all open?",
        "What are your business hours?",
        "When are you open for business?",
        "What are your opening hours?",
        "What time do you open?",
        "What time do you start your business day?",
        "What are your operating hours?",
        "When can I visit your store/office?",
        "What are your store hours?",
        "What time do you open for customers?",
        "What is your business opening time?",
        "When are you typically open?",
        "What are your hours of operation?",
        "When can I expect you to be open?",
        "What time does your business open?",
        "When is the best time to visit?",
        "What are your opening times?",
        "When can I come in?",
        "Are you open now?",
        "What are your hours of service?",
        "When can I expect to find you open?"
      ],
      "responses": [
        "We will be open from 8am - 9pm",
        "Our opening hours will be 8am to 9pm.",
        "We will be open from 8am until 9pm.",
        "We are open from 8am to 9pm.",
        "Our hours of operation are 8am to 9pm.",
        "We will be open between 8am and 9pm.",
        "You can visit us from 8am to 9pm.",
        "We will be available from 8am to 9pm.",
        "Our business hours are 8am to 9pm.",
        "We are open daily from 8am to 9pm.",
        "Our store will be open from 8am until 9pm.",
        "We will be open for business from 8am to 9pm.",
        "Our opening hours today are 8am to 9pm.",
        "We will be open all day from 8am to 9pm.",
        "You can reach us from 8am to 9pm.",
        "We are open for service from 8am to 9pm.",
        "Our operating hours today are 8am to 9pm.",
        "We will be available for service from 8am to 9pm.",
        "You can visit us between 8am and 9pm.",
        "We will be open during these hours: 8am to 9pm.",
        "Our store hours today are 8am to 9pm."
      ]
    }
  ]
}
'''

# print(data)

intents = json.loads(data)

def dataPreprocessing():
  # Data required for training
  tags = []
  all_words = []
  xy = []


  for intent in intents['intents']:
    # Storing the tags
    tag = intent['tag']
    tags.append(tag)

    for pattern in intent["patterns"]:
      # tokenizing the words from intents
      tokenized_word = tokenizer(pattern)

      all_words.extend(tokenized_word)
      xy.append((tokenized_word, tag))

  # stem and lower each word
  ignore_words = ['?', '.', '!']
  all_words = [stemming(w) for w in all_words if w not in ignore_words]

  # remove duplicates and sort
  all_words = sorted(set(all_words))
  tags = sorted(set(tags))

  # getting the x_train (e.g. [0, 1, 0, 1, 1]) and y_train (e.g. "greeting" which will be in index number like 1)
  x_train = []
  y_train = []

  print ("inside bag of words call")
  for (sentence, tag) in xy:
    bag = bag_of_words(sentence, all_words)
    x_train.append(bag)
    y_train.append(tags.index(tag))


  x_train = np.array(x_train)
  y_train = np.array(y_train)

  class ChatDataset(Dataset):
    def __init__(self):
      self.n_samples = len(x_train)
      self.x_data = x_train
      self.y_data = y_train

    def __getitem__(self, index):
      return self.x_data[index], self.y_data[index]

    def __len__(self):
      return self.n_samples

  # preparing a dataset basd on the json data
  dataset = ChatDataset()

  return dataset, x_train, y_train, tags, all_words

```

# Model and Training

```python
# Creating the model class
class neuralNetwork(nn.Module):
  def __init__(self, input_size, output_size, hidden_size):

    # initializing the constructor of the parent class
    super(neuralNetwork, self).__init__()

    # creating layers of the model as attibutes
    self.layer1 = nn.Linear(input_size, hidden_size)
    self.layer2 = nn.Linear(hidden_size, hidden_size)
    self.layer3 = nn.Linear(hidden_size, output_size)
    self.relu = nn.ReLU()

    # forward propogation
  def forward(self, x):
    out = self.layer1(x)
    out = self.relu(out)
    out = self.layer2(out)
    out = self.relu(out)
    out = self.layer3(out)

    return out


# Implementing the data preprocessing
(dataset, x_train, y_train, tags, all_words) = dataPreprocessing()


# trainloader helps iterate the dataset into batches
train_Loader = DataLoader(dataset=dataset, batch_size=8, shuffle= True, num_workers=2)

# creating the model
model= neuralNetwork(input_size=len(x_train[0]), output_size=len(tags), hidden_size=8)

training_loss = []

  # training the model
def train_model(model, train_Loader, epochs):
  # Loss function and optimizer
  criterion = nn.CrossEntropyLoss()
  optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

  # training the model
  for epoch in range(epochs):
    for (words, label) in train_Loader:
      # forward propogation
      output = model(words)
      loss = criterion(output, label)

      # Backward and optimizing
      # Zeroing the gradient before optimizing
      optimizer.zero_grad()


      # calcute the loss/gradient
      loss.backward()

      # optimize the parameters based on gradient
      optimizer.step()


# printing the progress of epochs
    if (epoch +1) % 100 == 0:
      print(f"epoch {epoch+1}/{epochs}, loss={loss.item():.4f}")
      training_loss.append(loss.item())

  print(f"final loss, loss={loss.item():.4f}")
  plt.plot(training_loss, label = "training loss")
  plt.legend()
  plt.title("loss over epochs")
  plt.show()

train_model(model, train_Loader, epochs=1500)
```

# Using the model and running the chatbot

But it has various limitations such as not answering the specific query but only a generalized response based on the category, does not remember the previous chat context, etc. In addition to this, spelling errors and slang terms also significantly reduces the accuracy by almost 20%.
